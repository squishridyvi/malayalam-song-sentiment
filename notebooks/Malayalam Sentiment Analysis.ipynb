```python
#2020s songs

## Angu vaana konilu 

file_path = os.path.join(html_folder, "Angu Vaana Konilu [2024].html")

with open(file_path, "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

# View the first part of the HTML
print(soup.prettify()[:])
```

```python
angu_vaana_konilu_div = soup.find("div", class_="nonselectable")

# get the inner div
angu_vaana_konilu_inner_div = angu_vaana_konilu_div.find("div")

# convert <br/> breaks into line breaks, extract the text
angu_vaana_konilu_lyrics_text = angu_vaana_konilu_inner_div.get_text(separator="\n").strip()

print(angu_vaana_konilu_lyrics_text)
```

## Creating a .csv file for manual labelling of sentiment 

```python
new_list = []

manual_sentiment = { 
    "Aangu Vaana Konilu": "Positive", 
    "Minnal Vala": "Positive", 
    "Eyy Banane": "Negative", 
    "Entammede Jimmiki Kammal": "Negative",
    "Malare Ninne": "Positive",
    "Kudukku": "Negative", 
    "Picha Vaacha Naal": "Positive", 
    "Nee oru puzhayaay": "Positive", 
    "Mazhavil": "Negative", 
    "Paathiraavaayi": "Neutral", 
    "Kilukilppambaram": "Negative", 
    "Saarangi maaril": "Positive", 
    "Doore Kizhakkudikkin": "Positive", 
    "Thumbi vaa": "Neutral", 
    "Vellichillum": "Neutral"
}

# appending the song name, the lyrics, and the manual sentiment all into one list 
for key, value in list_of_songs.items(): 
    sentiment = manual_sentiment[key]
    new_list.append({"song_name": key, "lyrics_text": value, "sentiment": sentiment})

# turning this new list into a table and csv files
sentiment_table = pd.DataFrame(new_list).to_csv('Sentiment_of_Malayalam_songs.csv', index=False)
```

```python
##translating the 2020s songs

song_name = "Aangu Vaana Konilu"
prompt = f"Translate this text from an f'string, and only write its translation:\n\n{list_of_songs['Aangu Vaana Konilu']}"
print(f"Translating: {song_name}")
print(prompt)
    
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": prompt}
] 
    
chat = client.chat.completions.create(
    model="gpt-4-turbo", 
    messages=messages
)

reply = chat.choices[0].message.content 
print(reply)

# add translated lyrics columns
translated_lyrics_df = pd.read_csv('Sentiment_of_Malayalam_songs.csv') 
cleaned_reply = re.sub(r'[^\x00-\x7F\n.,!?;:()\'" -]+', '', reply).strip()
translated_lyrics_df.loc[translated_lyrics_df['song_name'] == "Aangu Vaana Konilu", 'translated_lyrics'] = cleaned_reply
## add to original csv file 
translated_lyrics_df.to_csv('Sentiment_of_Malayalam_songs.csv', index=False)
```

## Using ChatGPT to classify the sentiments 

```python
import pandas as pd 

# for each song in the list given, I need it to classify it as positive, neutral or negative, and add that sentiment value to my csv file 
# under a column titled 'ChatGPT sentiment'

# reading the Sentiment of Malayalam songs csv file into memory 
chatgpt_df = pd.read_csv('Sentiment_of_Malayalam_songs.csv')     
for index, row in chatgpt_df.iterrows():
    prompt = f"Identify the overall sentiment of these English-translated Malayalam lyrics, and classify them as positive, neutral or negative. Only respond with the sentiment label \n\n{row['translated_lyrics']}"
    print(f"Translating: {row['song_name']}")
    print(prompt)
            
    messages = [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": prompt}
    ] 
            
    chat = client.chat.completions.create(
        model="gpt-4-turbo", 
        messages=messages
    )
        
    reply = chat.choices[0].message.content.strip()
    print(reply)
    
    chatgpt_df.at[index, 'ChatGPT sentiment'] = reply
## add to original csv file 
chatgpt_df.to_csv('Sentiment_of_Malayalam_songs.csv', index=False)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 
```

```python
# instance of vader_sentiment
vader_sentiment = SentimentIntensityAnalyzer() 

vader_sentiments_df = pd.read_csv('Sentiment_of_Malayalam_songs.csv') 

for index, row in vader_sentiments_df.iterrows(): 
    # accessing the translated lyrics for each song
    sentiment_score = vader_sentiment.polarity_scores(row['translated_lyrics'])
    if sentiment_score['compound'] >= 0.05: 
        label = "Positive" 
    elif sentiment_score['compound'] <= -0.05: 
        label = "Negative" 
    else: 
        label = "Neutral" 

    vader_sentiments_df.at[index, 'VADER sentiment'] = label
## add to original csv file 
vader_sentiments_df.to_csv('Sentiment_of_Malayalam_songs.csv', index=False)
```

```python
textblob_sentiments_df = pd.read_csv('Sentiment_of_Malayalam_songs.csv') 

for index, row in textblob_sentiments_df.iterrows(): 
    textblob_sentiment = TextBlob(row['translated_lyrics']) 

    textblob_sentiment_score = textblob_sentiment.polarity

    if textblob_sentiment_score > 0: 
        label = "Positive"
    elif textblob_sentiment_score < 0: 
        label = "Negative"
    else:
        label = "Neutral"

    textblob_sentiments_df.at[index, 'TextBlob sentiment'] = label

## add to original csv file 
textblob_sentiments_df.to_csv('Sentiment_of_Malayalam_songs.csv', index=False)
```
